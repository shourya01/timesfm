```python
import torch
from anytree import Node, RenderTree
from anytree.exporter import UniqueDotExporter

def state_dict_to_tree(state_dict, root_name='state_dict'):
    """
    Converts a PyTorch state_dict into a tree structure for visualization.

    Args:
        state_dict (dict): The state dictionary from a PyTorch model.
        root_name (str): The name of the root node in the tree.

    Returns:
        Node: The root node of the constructed tree.
    """
    root = Node(root_name)
    nodes = {root_name: root}

    for key, value in state_dict.items():
        parts = key.split('.')
        parent = root
        path = root_name
        for i, part in enumerate(parts):
            path = f"{path}.{part}" if path else part
            if path not in nodes:
                if i == len(parts) - 1:
                    # Leaf node: include tensor shape
                    node_name = f"{part}: Tensor, shape={tuple(value.shape)}"
                else:
                    node_name = part
                nodes[path] = Node(node_name, parent=parent)
            parent = nodes[path]
    return root
```

```python
state_dict = torch.load('/home/shourya01/timesfm/timesfm-1.0-200m-pytorch/torch_model.ckpt')
```

```python
tree_root = state_dict_to_tree(state_dict)

for pre, _, node in RenderTree(tree_root):
    print(f"{pre}{node.name}")
```
